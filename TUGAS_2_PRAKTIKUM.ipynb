```python
#Brand dan Model yang diminati
top_brands = df['Brand'].value_counts().nlargest(10).index
top_models = df['Model'].value_counts().nlargest(10).index

filtered_df = df[df['Brand'].isin(top_brands) & df['Model'].isin(top_models)]

plt.figure(figsize=(15, 10))
ax = sns.countplot(y='Brand', hue='Model', data=filtered_df, palette='viridis')

for p in ax.patches:
    ax.annotate(f'{p.get_width()}', (p.get_x() + p.get_width() + 0.1, p.get_y() + p.get_height()/2), ha='center')

plt.title('Top Luxury Watch Brands and Models')
plt.xlabel('Count')
plt.ylabel('Brand')
plt.legend(title='Model', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
